# 2022 Free Agent Data

```python
FA_2022 = pd.read_csv('roster-resource-download-2022.csv')
FA_2022
```

## Dealing with no contract players

```python
salary_0_2022 = FA_2022.loc[(FA_2022['Total Salary'] <=0) |(FA_2022['Total Salary'].isna())] #THese guys got no cotracrs
salary_0_2022 #Players with no salalry, but 1 year, so likely minor league contracrs
```

```python
data_2022 = FA_2022.drop(salary_0_2022.index) #Ensures that the data frame has 2
data_2022
```

```python
print(FA_2022.shape)
print(salary_0_2022.shape)
print(data_2022.shape)
#As you can see the data_2022 df has gotten rid of players that had a contract of 0
```

```python
pitchers_2022 = data_2022.loc[(data_2022['Pos'] == 'SP') | (data_2022['Pos'] == 'RP')] #Dataframe of pitchers
pitchers_2022
```

## Position Players

```python
position_players_2022 = data_2022.drop(pitchers_2022.index)
position_players_2022
#Dataframe of the postion players
```

```python
position_players_2022.columns
```

**Remember, the 2022 free agents will have statistics from 2021, make sure you don't get confused by that**

```python
fg_pos_2021 =pyb.batting_stats(start_season = 2021, qual = 1) #2021 data becuase thats what 2022 free agents were graded on
```

```python
len(fg_pos_2021.columns)#320 columns in the fangrpahs data, definetly need to cut down on this
```

Merging

```python
fg_pos_2021.loc[fg_pos_2021['Name'] == "Carlos Correa"]['IDfg']
```

```python
position_players_2022.loc[position_players_2022['Name'] == 'Carlos Correa']['playerid']#We can see that the columns playerid abd IDfg are the same value, can join on it
```

```python
position_players_2022.head()#I would now like to determine which columns are the most important from this Fangrpahs roster data
#Name, pos, bats, thr, prev team, age,
```

```python
col_of_interest = ['Name', 'Pos', 'Bats', 'Thr', 'Prev Team', 'Age', 'Service Time', 'QO', 'Signing Team', 'Years', 'Total Salary', 'AAV', 'playerid'] #Not oncluding the projections and crowd sourced data
```

```python
pos_salary_2022 = position_players_2022[col_of_interest] #This is our FG roster data with what I deem to be important columns
```

###Selecting the appropiate columns from FG data

```python
fg_pos_2021.loc[fg_pos_2021['Team'] =='- - -'].head()#Notice that free agents in this data are getting '- - -' in the team columns, this is okay tho since our position player data has
```

```python
for i in fg_pos_2021.columns:
  print(i)
```

```python
fg_stats_columns_interest = ['IDfg',	'Season', 'Name', 'G', 'AB', 'PA', 'H', '1B', '2B', '3B', 'HR', 'R', 'RBI', 'BB', 'IBB', 'SB', 'AVG', 'BB%', 'K%', 'OBP', 'SLG', 'OPS', 'ISO', 'BABIP', 'wOBA', 'xwOBA', 'wRC+', 'BABIP',
'LD%','GB%','FB%',
'Soft%', 'Med%', 'Hard%', 'WPA', 'WAR']
#You could try one model with woba, and one without xwoba
```

```python
pos_stats_2022 = fg_pos_2021[fg_stats_columns_interest]
pos_stats_2022.head()
```

```python
len(pos_stats_2022.columns)
```

Merging

```python
pos_salary_2022.head()
```

```python
pos_stats_2022 = pos_stats_2022.rename(columns= {'IDfg': 'playerid'})
```

```python
pos_stats_2022
```

```python
train_2022_pos = pos_salary_2022.merge(pos_stats_2022, how= 'left', on= 'playerid') #This is the dataframe of the combined features
```

```python
train_2022_pos = train_2022_pos.drop(columns= 'Name_y')
train_2022_pos = train_2022_pos.rename(columns= {'Name_x': 'Name'})
```

```python
train_2022_pos.to_csv('train_2022_pos.csv')
```

## Pitchers

```python
pitch_2021_data = pyb.pitching_stats(2021, qual = 1) #Grab fangraphs data from the previous year compared to 2022
```

```python
pitchers_2022 = pitchers_2022[col_of_interest] #Select columns with relevant columns from earlier (same as pos players)
pitchers_2022
```

```python
pitching_cols = ['Name', 'IDfg', 'Season', 'W', 'L', 'SV', 'G', 'GS', 'IP', #These are all of the columns that proably are good for pictures
                 'K/9', 'BB/9', 'HR/9', 'BABIP', 'LOB%',
                 'H', 'BB', 'SO', 'WP', 'K%', 'BB%',
                 'AVG', 'WHIP', 'ERA', 'FIP', 'E-F', 'xFIP',
                 'GB/FB', 'LD%', 'GB%', 'FB%',
                 'Soft%', 'Med%', 'Hard%', 'WPA', 'WAR']
#SD (Shutdowns): Number of relief appearances worth 0.06 WPA or more.

#MD (Meltdowns): Number of relief appearances worth -0.06 or less.
```

```python
pitch_stats_2021 = pitch_2021_data[pitching_cols]
pitch_stats_2021
```

Merging Pitching Data

```python
pitch_stats_2021 = pitch_stats_2021.rename(columns={'IDfg':'playerid'})#rename the id col so they 2 dataframes can merge
pitch_stats_2021
```

```python
train_2022_pitch = pitchers_2022.merge(pitch_stats_2021, how='left', on ='playerid')
train_2022_pitch = train_2022_pitch.drop(columns= 'Name_y') #Drop extra name col
train_2022_pitch = train_2022_pitch.rename(columns= {'Name_x': 'Name'}) #Rename it to be more clear
train_2022_pitch
```

```python
train_2022_pitch.to_csv('train_2022_pitch.csv')
```
